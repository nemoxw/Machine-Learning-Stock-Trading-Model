```python
import json

RL_portfolio = []
LSTM_portfolio = []

with open("RL_portfolio.json", 'r') as f:
    RL_portfolio = json.load(f)

with open("LSTM_portfolio.json", 'r') as f:
    LSTM_portfolio = json.load(f)

print(len(RL_portfolio))
print(len(LSTM_portfolio))
```

```python
import matplotlib.pyplot as plt

plt.plot(RL_portfolio, color = 'red', label = 'Reinforcement Learning Protfolio Value')
plt.plot(LSTM_portfolio, color = 'blue', label = 'LSTM Protfolio Value')
plt.title('Protfolio Value of the RL and LSTM trading model')
plt.xlabel('Time (trading days)')
plt.ylabel ('$')
plt.legend ()
```
